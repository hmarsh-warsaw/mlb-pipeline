```python
import duckdb
conn = duckdb.connect('../data/mlb.db')
```

## Raw sample

```python
conn.execute('SELECT * FROM prop_lines LIMIT 20').df()
```

## What markets do we have?

```python
conn.execute('''
    SELECT market, COUNT(*) as rows
    FROM prop_lines
    GROUP BY market
    ORDER BY rows DESC
''').df()
```

## What bookmakers do we have?

```python
conn.execute('''
    SELECT bookmaker, COUNT(*) as rows
    FROM prop_lines
    GROUP BY bookmaker
    ORDER BY rows DESC
''').df()
```

## Hits lines for today

```python
conn.execute('''
    SELECT player_name, home_team, away_team, bookmaker, line, side, odds
    FROM prop_lines
    WHERE market = 'batter_hits'
    AND bookmaker = 'draftkings'
    AND side = 'Over'
    ORDER BY line DESC
''').df()
```

## Home run lines for today

```python
conn.execute('''
    SELECT player_name, home_team, away_team, bookmaker, line, side, odds
    FROM prop_lines
    WHERE market = 'batter_home_runs'
    AND bookmaker = 'draftkings'
    ORDER BY player_name
''').df()
```

## Line comparison across bookmakers (picks a player)

```python
conn.execute('''
    SELECT player_name, market, bookmaker, line, side, odds
    FROM prop_lines
    WHERE player_name = 'Aaron Judge'
    ORDER BY market, bookmaker, side
''').df()
```

## Games today

```python
conn.execute('''
    SELECT DISTINCT away_team || ' @ ' || home_team as matchup, game_date
    FROM prop_lines
    ORDER BY game_date
''').df()
```
